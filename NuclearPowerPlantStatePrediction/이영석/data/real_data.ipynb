### 한 열에 하나의 값만 존재하는 경우 확인

```python
# 중복 체크
def is_exist(a, b):
    if not a:
        return False
    else :
        for l in a:
            if [b for i, j in zip(l, b) if i == j]:
                return True
```

```python
all_single_values = []
            
print("30.csv")
for col_name in df_30.columns:
    values = np.unique(df_30[col_name])
    if(len(values)) == 1:
        print(values)
        if not is_exist(all_single_values, values):
            all_single_values.append(values)
        
print("")


print("1154.csv")
for col_name in df_1154.columns:
    values = np.unique(df_1154[col_name])
    if(len(values)) == 1:
        print(values)
        if not is_exist(all_single_values, values):
            all_single_values.append(values)
        
print("")


print("1168.csv")
for col_name in df_1168.columns:
    values = np.unique(df_1168[col_name])
    if(len(values)) == 1:
        print(values)
        if not is_exist(all_single_values, values):
            all_single_values.append(values)
        
print("")
```

```python
for single_values in all_single_values:
    print(single_values)
```

### 한 열에 여러개의 값이 존재하는 경우 확인

```python
all_multi_values = []
            
            
print("30.csv")
for col_name in df_30.columns:
    values = np.unique(df_30[col_name])
    if(len(values)) > 1:
        print(values)
        if not is_exist(all_multi_values, values):
            all_multi_values.append(values)
print("")


print("1154.csv")
for col_name in df_1154.columns:
    values = np.unique(df_1154[col_name])
    if(len(values)) > 1:
        print(values)
        if not is_exist(all_multi_values, values):
            all_multi_values.append(values)
print("")


print("1168.csv")
for col_name in df_1168.columns:
    values = np.unique(df_1168[col_name])
    if(len(values)) > 1:
        print(values)
        if not is_exist(all_multi_values, values):
            all_multi_values.append(values)
print("")
```

```python
for multi_values in all_multi_values:
    print(multi_values)
```

### 결과 분석

~~~

#### 문자열

### 실수가 아닌 데이터 정리

- 데이터는 정수, 문자열로 존재함
- 대부분 한 열에 하나의 데이터(정수 또는 문자열)만 존재하였음


- 반드시 한 열에 하나의 데이터로만 존재하는 데이터(변하지 않는 데이터)들은 의미가 없을 것 같음

- 한 열에 다른 데이터와 같이 존재하는 데이터(변하는 데이터)만 고려해야 할 것 같음
    - 혼자 존재하는 경우를 구별하기 어려울것 같으므로 혼자 존재하는 경우도 따로 고려해야 할 것 같음
