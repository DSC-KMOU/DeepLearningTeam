# 실제 데이터 찾기

결측치 및 실수가 아닌 다른 데이터를 가지는 실제 데이터 파일 찾기

- float형 데이터가 아닐 경우 기록
- 해당 파일이름으로 파일 생성 및 해당 열 모두 기록

```python
# 데이터셋 경로

import os
import numpy as np
import pandas as pd

data_dir = "./datasets"

train_dir = os.path.join(data_dir, "train")
test_dir = os.path.join(data_dir, "test")

train_files = sorted(os.listdir(train_dir))
test_files = sorted(os.listdir(test_dir))
```

```python
# 기록할 디렉토리 경로 및 생성

target_dir = "./real_datas"

target_train_dir = os.path.join(target_dir, "train")
target_test_dir = os.path.join(target_dir, "test")

if not os.path.isdir(target_train_dir) :
    os.mkdir(target_train_dir)
    
if not os.path.isdir(target_test_dir) :
    os.mkdir(target_test_dir)
```

```python
# train 데이터 기록

for file in train_files:
    # 탐색중인 파일 출력
    print(file)
    df = pd.read_csv(os.path.join(train_dir, file))

    # 첫 행 (행 별로 모두 같은 데이터타입이므로)
    first_data_row = df.iloc[0]

    idx_list = []

    # time 축 제외
    for col_idx, data in enumerate(first_data_row[1:]):
        # float data가 아니면 index 리스트에 저장
        # time축 제외했으니까 index+1
        if str(type(data)).find('float') == -1:
            idx_list.append(col_idx+1)

    # 리스트에 저장된 데이터가 있으면 기록
    if(idx_list):
        names = df.columns.values[idx_list]
        datas = df[names]
        datas.to_csv(os.path.join(target_train_dir, file), index=False)
```

```python
# test 데이터 기록

for file in test_files:
    # 탐색중인 파일 출력
    print(file)
    df = pd.read_csv(os.path.join(test_dir, file))

    # 첫 행 (행 별로 모두 같은 데이터타입이므로)
    first_data_row = df.iloc[0]

    idx_list = []

    # time 축 제외
    for col_idx, data in enumerate(first_data_row[1:]):
        # float data가 아니면 index 리스트에 저장
        # time축 제외했으니까 index+1
        if str(type(data)).find('float') == -1:
            idx_list.append(col_idx+1)

    # 리스트에 저장된 데이터가 있으면 기록
    if(idx_list):
        names = df.columns.values[idx_list]
        datas = df[names]
        datas.to_csv(os.path.join(target_test_dir, file), index=False)
```
